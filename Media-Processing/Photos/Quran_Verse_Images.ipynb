# Image-Processing [Quran Ayah]
Lets apply different image-processing functions from the images.py script on a Quran-Ayah photo from *everyayah.com* (Surah 100, verse 1).

```python
from images import *

import requests
from io import BytesIO

photo_path = 'https://everyayah.com/data/quranpngs/100_1.png'

if photo_path.count(':') or photo_path.startswith('https://'):
    response = requests.get(photo_path)
    photo_path = BytesIO(response.content)
    
img = Image.open(photo_path).convert('RGB')
img_comment = 'Image size: [%d x %d]'

display(img)
print(img_comment%img.size)
```

See above the Image of the Quran-verse extracted from everyayah.com, it seems to have a long-left white space, further observed by **inverting the photo's colors**:

```python
img_invert(img)
```

**Trimming the empty space** out from the image:

```python
trimmed_img = img_trim(img)

display(trimmed_img)
print(img_comment%trimmed_img.size)
```

Sweet! We can also **sharpen the image** a bit, given that our cropped version above looks a bit blurry.

```python
img_sharpen(trimmed_img, amount=10, as_img=True)
```

Alright, finally lets **resize the photo** by 6 times it's width:

```python
big_img = img_resize(trimmed_img, by_width = trimmed_img.size[0]*6)
display(big_img)
print(img_comment%big_img.size)
```
