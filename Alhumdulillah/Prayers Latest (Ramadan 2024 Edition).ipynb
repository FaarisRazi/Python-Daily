```python
from IPython.display import HTML # For the last function below
import urllib.request as ur, pandas as pd, numpy as np, os
import datetime
    
prayers_df = pd.read_html('https://www.khaleejtimes.com/prayer-time-uae/dubai')[0]
prayers_df
```

```python

# For file management and dataframe-loading:
today = datetime.datetime.today()#.strftime('%d-%m-%Y')
month_now = today.strftime('%B')

today_info = prayers_df[prayers_df[month_now] == today.day]
today_prayers = today_info.T[4:].rename(columns={today_info.index[0]:'times'}).rename_axis('prayers')

print("Left: Prayers-DF (head)\t\t||\t"+
      f"Right: Today's prayer-times ({today.strftime('%d-%m-%Y')})")
display(prayers_df.head(), today_prayers)

month_days_left = prayers_df.loc[today_info.index[0]:]

days_left = month_days_left.shape[0]-1
days_comment = '%d days left for '+month_now

if days_left:
    print(days_comment%days_left)
    display(month_days_left[1:])
else:
    print('Today is the last day '+days_comment.split()[-1][:-1]+'-day')
```

### Iqamah timings (from local masjid)

```python
iqamah = {i:pd.Timedelta(minutes=j) if j else np.nan 
          for i,j in zip(today_prayers.index, [20,0,20,20,5,20])}

print('Iqamah times (my local masjid):')
display(iqamah)

prayers_astime = pd.to_datetime(today_prayers.times, format='%I:%M %p')

today_prayers['iqamah'] = prayers_astime.values + today_prayers.index.map(iqamah.get).values
today_prayers['iqamah'] = today_prayers['iqamah'].dt.strftime('%I:%M %p')
today_prayers
```

### Add durations until the next prayer/"session"
For each row in *today_prayers*, we will find the difference in time-duration or "time until next prayer" per row. 

<u>Isha's *until_next*</u> below is the duration from **today's Isha until today's Fajr** - assuming the same Fajr-time for the next day.

```python
today_prayers['until_next'] = prayers_astime.diff().fillna(pd.Timedelta(seconds=0))
#Or: diff().dt.seconds.fillna(0).apply(duration, 0) # duration() from my repo: Python-Daily/Date & Time/duration

today_prayers.loc['Fajr', 'until_next'] = prayers_astime.iloc[::-5].diff()[-1]
today_prayers['until_next'] = today_prayers.until_next.apply(lambda x: str(x).split()[-1])
today_prayers['until_next'] = np.append(today_prayers.until_next[1:].values, 
                                        today_prayers.loc['Fajr', 'until_next'])
today_prayers
```

```python
import pandas as pd, numpy as np, os
import datetime

ramadan_2024_link = 'https://www.khaleejtimes.com/ramadan-prayer-time-table-uae'
ramadan_df = pd.read_html(ramadan_2024_link)[0]

ramadan_df.to_csv('Ramadan 2024 - Khaleej Times.csv', index=False)

ramadan_df.head()
```

```python

# .insert(4, 'new_col_name', tmp)
work_start = '7:00 AM'
shift_hours = 6


# Converting "time" based columns to datetime-objects (eg; Fajr time to time-object)
time_columns = ramadan_df.columns[ramadan_df.columns.tolist().index('Fajr'):]
for time_col in time_columns:
    ramadan_df[time_col] = pd.to_datetime(ramadan_df[time_col], format='%I:%M %p')

    
# Get Times for Sahoor, Iftaar, Taraweeh, Work-timings
fajr = ramadan_df.Fajr
sahoor = fajr - pd.Timedelta(hours=1)
iftaar = ramadan_df.Iftar
isha = ramadan_df.Isha

taraweeh_start = isha + pd.Timedelta(minutes = 20 + 10 + 5)
taraweeh_end = taraweeh_start + pd.Timedelta(minutes = 25)
fast_hours = isha - fajr

clockin = pd.to_datetime(work_start, format='%I:%M %p')
go_for_work = clockin - pd.Timedelta(minutes = 43)

clockout_reminder = clockin + pd.Timedelta(minutes = shift_hours*60 - 20)
clockout = clockout_reminder + pd.Timedelta(minutes = 20)

workout_zone = iftaar - pd.Timedelta(minutes = 75)
sleep = taraweeh_end + pd.Timedelta(minutes = 40)

next_sahoor = np.append(sahoor[1:], sahoor.values[-1]) + pd.Timedelta(days = 1)
sleep_hours = next_sahoor - sleep

# today_prayers['iqamah'] = prayers_astime.values + today_prayers.index.map(iqamah.get).values
# today_prayers['iqamah'] = today_prayers['iqamah'].dt.strftime('%I:%M %p')
```

```python
ramadan_df.drop('Imsak', axis=1, inplace=True)
ramadan_df.set_index('Ramadan', inplace=True)

new_cols = {
             'sahoor':sahoor, 'fast_hours':fast_hours,
             'go_for_work':go_for_work, 'clockin':clockin, 'clockout_reminder':clockout_reminder, 'clockout':clockout,
             'workout_zone':workout_zone,
             'taraweeh_start':taraweeh_start, 'taraweeh_end':taraweeh_end,
             'sleep':sleep, 'sleep_hours':sleep_hours
            }

for new_col_name, new_col in new_cols.items():
    ramadan_df[new_col_name] = new_col
    
ramadan_df.head()
```

```python
ramadan_df = ramadan_df[['Weekday', 'Month', 'fast_hours', 'sleep_hours',
                         'sahoor', 'Fajr', 'Sunrise',
                         'go_for_work', 'clockin', 'Dhuhr', 'clockout_reminder', 'clockout',
                         'Asr', 'workout_zone', 'Iftar',
                         'Isha', 'taraweeh_start', 'taraweeh_end',
                         'sleep']]

ramadan_schedule = ramadan_df.copy()
```

```python
time_cols = ramadan_df.columns[ramadan_df.columns.tolist().index('sahoor'):]

for tcol in time_cols:
    ramadan_schedule[tcol] = ramadan_schedule[tcol].dt.strftime('%I:%M %p')
    
ramadan_schedule.fast_hours = ramadan_schedule.fast_hours .apply(lambda x: str(x).split()[-1])
ramadan_schedule.sleep_hours = ramadan_schedule.sleep_hours .apply(lambda x: str(x).split()[-1])
```

```python
ramadan_schedule

melted_df = pd.melt(ramadan_df[[i for i in ramadan_schedule.columns if 'hours' not in i and 'Weekday' not in i]],
                    id_vars=['Month'], var_name='Event', value_name='Timing')

# Convert 'Date' column to datetime format
melted_df['Month'] = pd.to_datetime(melted_df['Month'], format='%d/%B/%Y')

# Sort the DataFrame based on 'Date' and 'Event'
melted_df.sort_values(by=['Month', 'Timing']).reset_index(drop=True)

ramadan_schedule_alarms = melted_df
```
